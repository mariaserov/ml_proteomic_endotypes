```python
def make_toy_dataset(df, prots, n=3000, path="Data", 
                     name="toy_dataset", hypertension=True):
    if hypertension is True: 
        df = df.loc[df["Systolic.blood.pressure..automated.reading.0.0"]>=float(140)]
    else:
        pass
    
    # Sample 3000 rows
    new_index = random.sample(df.index.tolist(), 3000)
    newrow_df = df.loc[new_index]
    
    cols = list(df.columns)
    nonprotcols = [x for x in cols if x not in prots]
    nonprotcols= nonprotcols[:-2]
    
    sampled_prot = random.sample(prots, 30)
    cols = nonprotcols+sampled_prot
    new_df = newrow_df[cols]
    
    return new_df
    
```
